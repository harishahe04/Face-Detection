### Face Detection - CNN using Dlib

The CNN Model with Dlib requires the need to GPU to run successfully. The pre trained model can be downloaded from the below link:

http://dlib.net/files/mmod_human_face_detector.dat.bz2

```python
import dlib
import cv2
import matplotlib.pyplot as plt
```

```python
#Check CUDA Availability
!nvcc --version  #Command Line Code 
```

```python
cnn_detector = dlib.cnn_face_detection_model_v1('/content/data/mmod_human_face_detector.dat')
```

### Image 1

```python
image_1 = cv2.imread('/content/images/British Royal family.jpg')

#Converting to RGB for displaying the image using Matplotlib Pyplot
image_1_RGB = cv2.cvtColor(image_1, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(15,10))
plt.imshow(image_1_RGB);
```

```python
detection = cnn_detector(image_1)
```

```python
#Number of faces detected
len(detection)
```

```python
#Ploting and displaying the bounding boxes for faces

for face in detection:

  left,top,right,bottom = face.rect.left(), face.rect.top(), face.rect.right(), face.rect.bottom()
  cv2.rectangle(image_1_RGB, (left,top), (right,bottom), (0,255,0), 2)

plt.figure(figsize=(10,8))
plt.imshow(image_1_RGB)
```

### Image 2
